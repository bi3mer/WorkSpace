# Get Files

```python
time   = open('./ForEachVSFor/bin/Debug/time.csv', 'r')
memory = open('./ForEachVSFor/bin/Debug/memory.csv', 'r')
```

# Turn Files into Lists

```python
def convert_files(f):
    # read
    content = f.read()
    
    # split by new line
    lines = content.split('\n')
    
    # remove last new line that is empty
    del lines[-1]
    
    # split each row by comma
    for i in range(len(lines)):
        lines[i] = lines[i].split(',')
        
    # convert strings into floats
    for i in range(len(lines)):
        for j in range(len(lines[i])):
            lines[i][j] = float(lines[i][j])
        
    return lines
```

```python
time_csv   = convert_files(time)
memory_csv = convert_files(memory)
```

# Create Graphs

## Import

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
```

## Time

```python
plt.xlabel("Number of Iterations (1e8)")
plt.ylabel("Time (S)")
plt.title("For vs. Foreach: Time to Run")

plt.plot(time_csv[0], time_csv[2], '-', label="Foreach Loop")
plt.plot(time_csv[0], time_csv[1], '-', label="For Loop")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

## Memory

```python
plt.plot(memory_csv[0], memory_csv[2], '-', label="Foreach Loop")
plt.plot(memory_csv[0], memory_csv[1], '-', label="For Loop")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
